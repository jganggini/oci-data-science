````python
import pypyodbc as odbc
import pandas as pd
````

````python
DRIVER_NAME = 'SQL SERVER'
SERVER_NAME = '10.100.***.*'
DATABASE_NAME = 'BD'
DATABASE_UID = 'admin'
DATABASE_PWD ='***'

connectionString = f'DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={SERVER_NAME};DATABASE={DATABASE_NAME};UID={DATABASE_UID};PWD={DATABASE_PWD};Trust_Connection=yes'
````

````python
conn = odbc.connect(connectionString)
print(conn)
````

````python
cursor  = conn.cursor()
cursor .execute("SELECT TOP 3 * FROM ENCUESTA_CLIMASAC")
result = cursor .fetchall()
````

````python
df = pd.DataFrame(result)
df
````
